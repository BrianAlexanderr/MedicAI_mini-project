```python
import pandas as pd
```

```python
df = pd.read_csv('./dataset_corrected.csv')
df.head()
```

```python
df = df.loc[:, ~df.columns.duplicated()]
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
import numpy as np

label_encoder = LabelEncoder()
```

```python
X = df.drop(columns=['prognosis'])
y = df['prognosis']
y_encoded = label_encoder.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
model = XGBClassifier(objective="multi:softprob", num_class=len(label_encoder.classes_), use_label_encoder=False, eval_metric="mlogloss")
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix

y_proba = model.predict_proba(X_test)

y_pred_encoded = np.argmax(y_proba, axis=1)  # Ambil kelas dengan probabilitas tertinggi
y_pred = label_encoder.inverse_transform(y_pred_encoded)  # Ubah ke nama penyakit

# 8. Evaluasi model
accuracy = accuracy_score(y_test, y_pred_encoded)
print(f"ðŸŽ¯ Akurasi Model: {accuracy:.4f}")

# 9. Contoh hasil prediksi
print("ðŸ”¹ Contoh prediksi penyakit:", y_pred[0])
print("ðŸ”¹ Probabilitas masing-masing penyakit:", y_proba[0])
```

```python
import pickle

file = open('model.pkl', 'wb')
pickle.dump(model, file)
file.close()

file = open('label_encoder.pkl', 'wb')
pickle.dump(label_encoder, file)
file.close()
```
